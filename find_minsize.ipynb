```
from __future__ import print_function, division
import pandas as pd
import numpy as np

from torch.utils.data import Dataset, DataLoader
import torch
import torch.nn as nn
import torch.optim as optim
import torch.utils.model_zoo as model_zoo
import torchvision
from torchvision import datasets, models, transforms
import matplotlib.pyplot as plt
import time
import os# dicom 이미지 읽기
import pydicom 
import tensorflow as tf
import tensorflow_io as tfio
from pydicom.data import get_testdata_file
from skimage import io, transform
```

```
folder = '/content/gdrive/My Drive/ISIC-melanoma/train/'

start_num = 10000
end_num = 10000000

min_size_height = 1000000   #이미지 최소 높이
min_size_height_fileNum = 0
min_size_weight = 1000000   #이미지 최소 너비
min_size_weight_fileNum = 0

mean_aspect_ratio = 0.00000   #이미지 가로세로 평균 비율
sum_aspect_ratio = 0.00000

mean_size_height = 0   #이미지 평균 높이
sum_size_height = 0

mean_size_weight = 0  #이미지 평균 너비
sum_size_weight = 0

file_startname = 'ISIC_00'
image_num  = 0

for i in range (start_num, end_num):
  if 1000000 > i >= 100000:
    file_startname = 'ISIC_0'
  elif 10000000 > i >= 1000000:
    file_startname = 'ISIC_'

  if os.path.isfile(folder + file_startname + str(i) + '.dcm'):
    image_num = image_num + 1
    filename = folder + file_startname + str(i) + '.dcm'
    dataset = pydicom.dcmread(filename)

    image_bytes = tf.io.read_file(filename)

    # image = tfio.image.decode_dicom_image(image_bytes, dtype=tf.uint16)

    # image_info = np.squeeze(image.numpy())

    rows = int(dataset.Rows)
    cols = int(dataset.Columns)

    if rows < min_size_height:
      min_size_height = rows
      min_size_height_fileNum = i

    if cols < min_size_weight:
      min_size_weight = cols
      min_size_weight_fileNum = i

    sum_size_height = sum_size_height + rows
    sum_size_weight = sum_size_weight + cols
    
    aspect_ratio = cols/rows
    sum_aspect_ratio = sum_aspect_ratio + aspect_ratio

    mean_aspect_ratio = sum_aspect_ratio / image_num
    mean_size_height = sum_size_height / image_num
    mean_size_weight = sum_size_weight / image_num

    print(image_num, ' : ', aspect_ratio, ' | 최소 높이 : ', min_size_height, ' | 최소 너비 : ', min_size_weight, ' | 평균 높이 : ', mean_size_height, ' | 평균 너비 : ', mean_size_weight, ' | 평균 비율 : ', mean_aspect_ratio)



    
```

```
print('평균 가로세로 비율 : ', mean_aspect_ratio)
print('평균 높이 : ', mean_size_height)
print('평균 너비 : ', mean_size_weight)
print('최소 높이 : ', min_size_height, ' | 해당 파일 번호 : ', min_size_height_fileNum)
print('최소 너비 : ', min_size_weight, ' | 해당 파일 번호 : ', min_size_weight_fileNum)
```
